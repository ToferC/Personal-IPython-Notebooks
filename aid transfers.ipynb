```python
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import re
import networkx as nx
import csv
import json

import plotly.plotly as py
import plotly.tools as tls
from plotly.graph_objs import *

mpl_fig = plt.figure()
```

```python
data_path = r"/Users/christopherallison/Documents/Coding/Data/"
```

```python
nodes = []
edges = []
details = {}

G = nx.DiGraph()
```

```python
# Read CSV of UID and Colleagues and create nodes & edges

with open(data_path+"aid_tables.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        edges.append((row['Source'], row['Target'], row['Weight']))
```

```python
for e in G.edges(data=True):
    print(e[2]['weight'])
```

```python
for e in edges:
    try:
        details[e[1]] =+ float(e[2])
    except IndexError:
        details[e[1]] = float(e[2])

for n in details.values():
    print(n)
```

```python
for k, v in details.items():
    try:
        G.add_node(k, aid_received=v)
    except KeyError:
        G.add_node(k, aid_received=0)
```

```python
for e in edges:
    G.add_edge(e[0], e[1], weight=e[2])
```

```python
for n in G.nodes():
    try:
        G.node[n]['aid_received'] == ""
    except KeyError:
        G.node[n]['aid_received'] = 0
```
