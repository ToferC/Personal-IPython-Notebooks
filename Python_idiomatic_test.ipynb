```python
nodes = []
edges = []
details = {}
errors = []
```

```python
# Read CSV of UID and names and create nodes for each user

with open(data_path+"gcc_user_info_update_2015-09-03.csv", "r") as f:
    reader = csv.reader(f, delimiter=',')
    next(reader)

    for row in reader:
        uid, email, start = row
        ampersand = email.find("@")
        name = email[0:ampersand]
        department = email[ampersand + 1:]
        details[uid] = {'name':name, 'department':department,
                        'start': start,'end': "01-09-2015"}
        nodes.append(row[0])

```

```python
# Read CSV blogs and create edges for connections

with open(data_path+"gccResults_blogs.csv", 'r') as f:
    reader = csv.reader(f, delimiter=',')
    next(reader)
    
    for row in reader:
        if len(row) > 1:
            blog_data = row[0].split(' ')
            blog, author, date_created, extra = blog_data
        
            for c in row[1:]: # Get comments (2+ columns)
                comment = c.split()
                blog_ref, commenter, comment_date = comment
                edges.append([commenter, author, comment_date])
            
```

```python
# Prepare edges

edge_dict = {}

for edge in edges:
    k = "{}, {}".format(edge[0], edge[1])
    edge_dict[k]['weight'] = edge_dict.get(k['weight'], {'weight': 0}) + 1
```
