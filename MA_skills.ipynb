```python
skills = df["Skill Name"]
```

```python
slist = [x for x in skills]
```

```python
slist
```

```python
# slist_tokens now breaks down each entry into nltk word tokens that can be manipulated.

# clean = only real words but allows only single word entries
slist_tokens_clean = [nltk.word_tokenize(x) for x in slist if wordnet.synsets(x)]

# raw = uncleaned, but splits each entry into a list.
slist_tokens_raw = [nltk.word_tokenize(x) for x in slist]
```

```python
def clean_up_text(list1):
    try:
        x = [w for w in list1 if w not in stopwords.words('english') and
            w.lower() not in stopwords.words('french') and wordnet.synsets(w)]
        if len(x):
            return x
    except AttributeError:
        raise
    #return(x)

    #fdist2 = nltk.FreqDist(x)
    #fdist2.plot(50, cumulative=True)
```

```python
clean_slist = [clean_up_text(item) for item in slist_tokens_raw]
```

```python
# Removed stopwords and non-existent words
clean_slist
```

```python
slist_text = nltk.Text(clean_slist)
```

```python
# slist_text is the cleaned list of lists, with each list representing a single entry into GCconnex
slist_text = [x for x in slist_text if x]
slist_text
```

```python
# Remove None entries
slist_corpus = [" ".join(x) for x in slist_text]
```

```python
# Set all responses to lower case for easier comparison.
slist_corpus_low = [x.lower() for x in slist_corpus]
```

```python
# slist_corpus represents the cleaned text reformatted into a set of strings for each entry and appropriate for NLTK analyis and manipulation.
len(set(slist_corpus_low))
```

### Take cleaned and combined text and create a frequency distribution on the top 50 results
