```python
total = defaultdict(float)

lil = {}
chris = {}
joint = {}
discard = {}

cards = {"'5191330007582439'": "Chris", "'5191330003766655'": "Liliane"}
```

```python
with open(os.path.join(budget_dir + "BMO Aug 2015.csv"), 'r+') as f:
    reader = csv.reader(f, delimiter = ',')
    next(reader)
    next(reader)
    next(reader)
    
    for line in reader:
        item, card, trans_date, post_date, value, details = line
        print('''Card: {card}
        Date: {post_date}
        Value: {value}
        Details: {details}
        '''.format(card=cards[card], post_date=post_date, value=value, details=details))
        allocate = input("Choose an payor (l, c, j or d to discard)")
        if allocate == 'c':
            total['chris'] += float(value)
            lil[item] = [card, trans_date, post_date, value, details]
        elif allocate == 'l':
            total['liliane'] += float(value)
            chris[item] = [card, trans_date, post_date, value, details]
        elif allocate == 'd':
            discart[item] = [card, trans_date, post_date, value, details]
        else:
            total['joint'] += float(value)
            joint[item] = [card, trans_date, post_date, value, details]
    print('Final Summary:')
    for k,v in total.items(): 
        print('{}: {}'.format(k, v))
            
    
```
