The code below will modify the column headings for the text to be lower case and to replace spaces with '_'.

### This makes it easier to use the column names in Python.

```python
def camel_to_snake(column_name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', column_name)
    return re.sub('(.)([a-z0-9])([A-Z])', r'\1_\2', s1).lower()

def strip_spaces(column_name):
    return column_name.replace(" ", "_")
```

```python
camel_to_snake("Java Loves Camel Case")
```

```python
df.columns = [camel_to_snake(col) for col in df.columns]
df.columns = [strip_spaces(col) for col in df.columns]
df.columns.tolist()
```

The code below goes through the items and looks for numerical groups and then pulls those into a new column - q23_level_clean.

```python
def reformat_string(value):
    try:
        x = int(re.search(r'\d+', value).group())
    except:
        ValueError
        x = 0
    return x
```

```python
df['q23_level_clean'] = [reformat_string(x) for x in df.q23_level]
del df['q23_level']
```

```python
def create_awesome_crosstabs(cross, columns):
    # Create cross-tabs by two lists of columns
            
    for column in columns:
        for item in column:
            x = pd.crosstab(cross, item)

            newfilename = '{}.csv'.format(item.name)
            x.to_csv(os.path.join(root_path, 'group_level/', newfilename))
```
